```python
best_quarters_perf=pd.DataFrame(data=ret)
```

```python
#shifting up&down the data in order to get the quarters adjacent to the ones we analyze
best_quarters_perf['Prev']=best_quarters_perf['Close'].shift(1)
best_quarters_perf['Next']=best_quarters_perf['Close'].shift(-1)
```

```python
#in order to avoid a nan value be part of the list, we need to fill the empty cells with 0.
best_use=best_quarters_perf.fillna(0).sort_values(by='Close').tail(10)
best_use
```

```python
#we want to have the order be previos,current,next quarter

columns = best_use.columns.tolist()

# Find the indices of the columns
index1 = columns.index('Close')
index2 = columns.index('Prev')

# Swap the positions of the columns
columns[index1], columns[index2] = columns[index2], columns[index1]

# Reassign the modified column order to the DataFrame
best_use = best_use[columns]
```

```python
#we want to have all the quarters and fill only the values that correspond to the quarters we need.
index_df = pd.DataFrame(best_quarters_perf.index)

merged_df = pd.merge(index_df, best_use[['Close']], left_on='Date', right_index=True, how='left')
merged_df.set_index('Date',inplace=True)
```

```python
#after that we can shift those values and drop the nan rows in order to get the index values of all the quarters that are left
best_quarters_perf.drop(merged_df.shift(1).dropna().index,inplace=True)
best_quarters_perf.drop(merged_df.shift(-1).dropna().index,inplace=True)
best_quarters_perf.drop(best_use.index,inplace=True)
```

```python
fig=go.Figure()
for col in best_use.columns:
    fig.add_trace(go.Box(name=col,y=best_use[col]*100,boxmean=True))
fig.add_trace(go.Box(name='Remaining',y=best_quarters_perf['Close']*100,boxmean=True))
fig.add_trace(go.Scatter(name='2023-03',marker=dict(color='black'),mode='markers',x=best_use.columns,y=best_use.loc['2023-03'].values[0]*100))
fig.update_layout(title='Best 10 Quarters Performance')
fig.show()
#the previous quarter is typically not anything spectacular.
```

```python
#are there any corrections in the next quarter?
```

```python
hmm=(1+ndx.loc[halves_use.index[2]:f'{halves_use.index[2].year}']['Close'])
hmm=hmm.to_frame()
hmm['Max']=hmm.apply(lambda x: max(hmm.loc[:x.name,'Close']),axis=1)
hmm['Min']=hmm.apply(lambda x: min(hmm.loc[:x.name,'Close']),axis=1)
hmm['Pct']=(hmm['Close']-hmm['Max'])/hmm['Max']

new=hmm.loc[hmm[hmm['Pct']<-0.05].iloc[0].name:]
dates=[]
dates.append(hmm[hmm['Pct']<-0.05].iloc[0].name)
i=1
#since we know that at a certain a point our while loop will run into an error where there is no such index value to match
#the filtering we have. at that point the loop breaks.
while i>0:
    try:
        #not a perfect solution. it only acknowledges when the correction starts and
        #then it automatically start searching for an upshot of 10% or more.
        #technically since we know the dates of the correction start, we can look for the recent high and evaluate the drop
        #also we can establish when the trough occurs since we know the date when the market has rebounded 10% or more.
        new=new.loc[new[new['Pct']<-0.05].iloc[0].name:]
        new['Min']=new.apply(lambda x: min(new.loc[:x.name,'Close']),axis=1)
        new['Pct']=(new['Close']-new['Min'])/new['Min']
        new=new.loc[new[new['Pct']>0.05].iloc[0].name:]
        new['Max']=new.apply(lambda x: max(new.loc[:x.name,'Close']),axis=1)
        new['Pct']=(new['Close']-new['Max'])/new['Max']
        dates.append(new[new['Pct']<-0.05].iloc[0].name)
        new=new.loc[new[new['Pct']<-0.05].iloc[0].name:]
        i+=1
    except IndexError:
        break
```

```python
#applying this logic to all the second halves after the top 10 first halves

num=np.empty(0)
correct=0.1
for half in halves_use.index:
    hmm=(1+ndx.loc[half:f'{half.year}']['Close'])
    hmm=hmm.to_frame()
    hmm['Max']=hmm.apply(lambda x: max(hmm.loc[:x.name,'Close']),axis=1)
    hmm['Min']=hmm.apply(lambda x: min(hmm.loc[:x.name,'Close']),axis=1)
    hmm['Pct']=(hmm['Close']-hmm['Max'])/hmm['Max']

    if min(hmm['Pct'])>-0.05:
        num=np.append(num,0)
    else:
        new=hmm.loc[hmm[hmm['Pct']<-0.05].iloc[0].name:]
        dates=[]
        dates.append(hmm[hmm['Pct']<-0.05].iloc[0].name)
        i=1
        #since we know that at a certain a point our while loop will run into an error where there is no such index value to match
        #the filtering we have. at that point the loop breaks.
        while i>0:
            try:
                #not a perfect solution. it only acknowledges when the correction starts and
                #then it automatically start searching for an upshot of 10% or more.
                #technically since we know the dates of the correction start, we can look for the recent high and evaluate the drop
                #also we can establish when the trough occurs since we know the date when the market has rebounded 10% or more.
                new=new.loc[new[new['Pct']<-correct].iloc[0].name:]
                new['Min']=new.apply(lambda x: min(new.loc[:x.name,'Close']),axis=1)
                new['Pct']=(new['Close']-new['Min'])/new['Min']
                new=new.loc[new[new['Pct']>correct].iloc[0].name:]
                new['Max']=new.apply(lambda x: max(new.loc[:x.name,'Close']),axis=1)
                new['Pct']=(new['Close']-new['Max'])/new['Max']
                dates.append(new[new['Pct']<-correct].iloc[0].name)
                new=new.loc[new[new['Pct']<-correct].iloc[0].name:]
                i+=1
            except IndexError:
                num=np.append(num,i)
                break
            
```
